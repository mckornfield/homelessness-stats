# 06 — Income & Poverty vs. Homeless Rate

Examines relationships between homelessness and economic conditions by state:
- **Median household income** (Census ACS 2022)
- **Poverty rate** (Census ACS 2022)
- **Gini coefficient** (Census ACS 2022 — income inequality)


```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_state_data.csv')
df = df.dropna(subset=['homeless_rate_per_10k', 'median_income', 'poverty_rate', 'gini_coefficient'])
print(f'Loaded {len(df)} states')
```

```python
slope, intercept, r, p, se = stats.linregress(df['median_income'], df['homeless_rate_per_10k'])
r2 = r ** 2
print(f'Median income vs rate: r={r:.3f}, R²={r2:.3f}, p={p:.4f}')

x_range = [df['median_income'].min(), df['median_income'].max()]
y_range = [slope * x + intercept for x in x_range]

fig1 = px.scatter(
    df, x='median_income', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Median Household Income vs. Homeless Rate by State<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f} (Census ACS 2022)</sup>',
    labels={'median_income': 'Median Household Income ($)', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig1.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig1.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig1.show()
```

```python
slope2, intercept2, r2v, p2, se2 = stats.linregress(df['poverty_rate'], df['homeless_rate_per_10k'])
r2_2 = r2v ** 2
print(f'Poverty rate vs homeless rate: r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f}')

x_range2 = [df['poverty_rate'].min(), df['poverty_rate'].max()]
y_range2 = [slope2 * x + intercept2 for x in x_range2]

fig2 = px.scatter(
    df, x='poverty_rate', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Poverty Rate vs. Homeless Rate by State<br><sup>r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f} (Census ACS 2022)</sup>',
    labels={'poverty_rate': 'Poverty Rate (%)', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig2.add_trace(go.Scatter(x=x_range2, y=y_range2, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig2.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig2.show()
```

```python
slope3, intercept3, r3v, p3, se3 = stats.linregress(df['gini_coefficient'], df['homeless_rate_per_10k'])
r2_3 = r3v ** 2
print(f'Gini vs homeless rate: r={r3v:.3f}, R²={r2_3:.3f}, p={p3:.4f}')

x_range3 = [df['gini_coefficient'].min(), df['gini_coefficient'].max()]
y_range3 = [slope3 * x + intercept3 for x in x_range3]

fig3 = px.scatter(
    df, x='gini_coefficient', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Gini Coefficient (Income Inequality) vs. Homeless Rate by State<br><sup>r={r3v:.3f}, R²={r2_3:.3f}, p={p3:.4f} (Census ACS 2022)</sup>',
    labels={'gini_coefficient': 'Gini Coefficient', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig3.add_trace(go.Scatter(x=x_range3, y=y_range3, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig3.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig3.show()
```
