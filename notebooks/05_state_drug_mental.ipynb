# 05 — Drug Use & Mental Health vs. Homeless Rate

Tests correlations between state-level homelessness and:
- **Drug use disorder** (SAMHSA NSDUH 2021-2022)
- **Any Mental Illness (AMI)** prevalence (SAMHSA NSDUH 2021-2022)
- **Overdose death rate** (CDC WONDER, per 100k)

```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_state_data.csv')
df = df.dropna(subset=['homeless_rate_per_10k', 'drug_disorder_pct', 'ami_pct', 'overdose_rate_per_100k'])
print(f'Loaded {len(df)} states')
```

```python
slope, intercept, r, p, se = stats.linregress(df['drug_disorder_pct'], df['homeless_rate_per_10k'])
r2 = r ** 2
print(f'Drug disorder vs rate: r={r:.3f}, R²={r2:.3f}, p={p:.4f}')

x_range = [df['drug_disorder_pct'].min(), df['drug_disorder_pct'].max()]
y_range = [slope * x + intercept for x in x_range]

fig1 = px.scatter(
    df, x='drug_disorder_pct', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Drug Use Disorder Rate vs. Homeless Rate by State<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f} (SAMHSA NSDUH 2021-2022)</sup>',
    labels={'drug_disorder_pct': 'Drug Use Disorder (% adults)', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig1.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig1.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig1.show()
```

```python
slope2, intercept2, r2v, p2, se2 = stats.linregress(df['ami_pct'], df['homeless_rate_per_10k'])
r2_2 = r2v ** 2
print(f'AMI vs rate: r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f}')

x_range2 = [df['ami_pct'].min(), df['ami_pct'].max()]
y_range2 = [slope2 * x + intercept2 for x in x_range2]

fig2 = px.scatter(
    df, x='ami_pct', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Any Mental Illness (AMI) Prevalence vs. Homeless Rate by State<br><sup>r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f}</sup>',
    labels={'ami_pct': 'AMI Prevalence (% adults 18+)', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig2.add_trace(go.Scatter(x=x_range2, y=y_range2, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig2.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig2.show()
```

```python
slope3, intercept3, r3v, p3, se3 = stats.linregress(df['overdose_rate_per_100k'], df['homeless_rate_per_10k'])
r2_3 = r3v ** 2
print(f'Overdose rate vs homeless rate: r={r3v:.3f}, R²={r2_3:.3f}, p={p3:.4f}')

x_range3 = [df['overdose_rate_per_100k'].min(), df['overdose_rate_per_100k'].max()]
y_range3 = [slope3 * x + intercept3 for x in x_range3]

fig3 = px.scatter(
    df, x='overdose_rate_per_100k', y='homeless_rate_per_10k', text='state_postal',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title=f'Drug Overdose Death Rate vs. Homeless Rate by State<br><sup>r={r3v:.3f}, R²={r2_3:.3f}, p={p3:.4f} (CDC WONDER)</sup>',
    labels={'overdose_rate_per_100k': 'Overdose Deaths per 100k', 'homeless_rate_per_10k': 'Homeless Rate per 10k'},
    hover_name='state_name',
)
fig3.add_trace(go.Scatter(x=x_range3, y=y_range3, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig3.update_traces(textposition='top center', selector=dict(mode='markers+text'))
fig3.show()
```
