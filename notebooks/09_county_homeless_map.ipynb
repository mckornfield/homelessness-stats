# 09 — County/CoC Homeless Rate Map

Choropleth map of homeless rate per 10,000 residents by county (CoC primary county FIPS).
Also shows top/bottom 15 CoCs by homeless rate.


```python
from pathlib import Path
import pandas as pd
import plotly.express as px

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_county_data.csv')
df['county_fips'] = df['county_fips'].astype(str).str.zfill(5)
print(f'Loaded {len(df)} CoCs')
```

```python
fig = px.choropleth(
    df,
    geojson='https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json',
    locations='county_fips',
    color='homeless_rate_per_10k',
    scope='usa',
    color_continuous_scale='Reds',
    hover_name='coc_name',
    hover_data={'county_fips': False, 'homeless_rate_per_10k': ':.1f', 'total_homeless': ':,', 'unsheltered_pct': ':.1f'},
    title='Homeless Rate per 10,000 Residents by County/CoC (HUD 2023)',
    labels={'homeless_rate_per_10k': 'Rate per 10k'},
)
fig.show()
```

```python
top15 = df.nlargest(15, 'homeless_rate_per_10k')
bot15 = df.nsmallest(15, 'homeless_rate_per_10k')
combined = pd.concat([top15, bot15])

fig2 = px.bar(
    combined.sort_values('homeless_rate_per_10k', ascending=True),
    x='homeless_rate_per_10k', y='coc_name', orientation='h',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title='Top 15 and Bottom 15 CoCs: Homeless Rate per 10k (2023)',
    labels={'homeless_rate_per_10k': 'Rate per 10k', 'coc_name': ''},
)
fig2.update_layout(height=700)
fig2.show()
```
