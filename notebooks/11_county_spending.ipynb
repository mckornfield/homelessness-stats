# 11 — Spending per Homeless vs. Homeless Rate (County/CoC)


```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_county_data.csv')
df = df.dropna(subset=['homeless_rate_per_10k', 'spending_per_homeless'])
df = df[df['spending_per_homeless'] > 0]
print(f'Loaded {len(df)} CoCs')
```

```python
slope, intercept, r, p, se = stats.linregress(df['spending_per_homeless'], df['homeless_rate_per_10k'])
r2 = r ** 2
print(f'Spending/homeless vs rate: r={r:.3f}, R²={r2:.3f}, p={p:.4f}')

x_range = [df['spending_per_homeless'].min(), df['spending_per_homeless'].max()]
y_range = [slope * x + intercept for x in x_range]

fig = px.scatter(
    df, x='spending_per_homeless', y='homeless_rate_per_10k',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    hover_name='coc_name',
    title=f'HUD CoC Spending per Homeless Person vs. Homeless Rate (CoC Level)<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f}</sup>',
    labels={'spending_per_homeless': 'HUD Spending per Homeless ($)', 'homeless_rate_per_10k': 'Rate per 10k'},
)
fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
fig.show()
```
