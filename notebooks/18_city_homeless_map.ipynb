# 18 — City/CoC Homeless Rate Bubble Map

Bubble map of homeless rate per 10k by CoC centroid (lat/lon from HUD CoC shapefiles).
Bubble size = total homeless count; bubble color = rate per 10k.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_city_data.csv')
df = df.dropna(subset=['lat', 'lon'])
print(f'Loaded {len(df)} CoCs with coordinates')
```

```python
fig = px.scatter_geo(
    df, lat='lat', lon='lon',
    size='total_homeless',
    color='homeless_rate_per_10k',
    color_continuous_scale='Reds',
    scope='usa',
    hover_name='city_name',
    hover_data={'lat': False, 'lon': False, 'state_postal': True, 'total_homeless': ':,', 'homeless_rate_per_10k': ':.1f', 'unsheltered_pct': ':.1f'},
    title='Homeless Rate per 10k by Major City/CoC (HUD 2023) — bubble size = total homeless',
    labels={'homeless_rate_per_10k': 'Rate per 10k'},
    size_max=60,
)
fig.update_layout(geo_bgcolor='lightblue')
fig.show()
```

```python
top20 = df.nlargest(20, 'homeless_rate_per_10k')
fig2 = px.bar(
    top20.sort_values('homeless_rate_per_10k', ascending=True),
    x='homeless_rate_per_10k', y='city_name', orientation='h',
    color='homeless_rate_per_10k', color_continuous_scale='Reds',
    title='Top 20 CoCs by Homeless Rate per 10k (2023)',
    labels={'homeless_rate_per_10k': 'Rate per 10k', 'city_name': ''},
)
fig2.show()
```
