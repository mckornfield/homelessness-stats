# 12 — Drug Use & Mental Health vs. Homeless Rate (County/CoC)

Uses state-level SAMHSA proxies joined to each CoC by state.


```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_county_data.csv')
df = df.dropna(subset=['homeless_rate_per_10k', 'ami_pct', 'drug_disorder_pct'])
print(f'Loaded {len(df)} CoCs')
```

```python
slope, intercept, r, p, se = stats.linregress(df['ami_pct'], df['homeless_rate_per_10k'])
r2 = r ** 2
print(f'Any Mental Illness (AMI) vs rate: r={r:.3f}, R²={r2:.3f}, p={p:.4f}')

x_range = [df['ami_pct'].min(), df['ami_pct'].max()]
y_range = [slope * x + intercept for x in x_range]

fig = px.scatter(
    df, x='ami_pct', y='homeless_rate_per_10k', color='state_postal',
    hover_name='coc_name',
    title=f'AMI Prevalence (State Proxy) vs. Homeless Rate (CoC Level)<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f} — SAMHSA NSDUH 2021-2022</sup>',
    labels={'ami_pct': 'AMI Prevalence (% adults, state proxy)', 'homeless_rate_per_10k': 'Rate per 10k'},
)
fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='black', dash='dash'), showlegend=False))
fig.show()
```

```python
slope2, intercept2, r2v, p2, se2 = stats.linregress(df['drug_disorder_pct'], df['homeless_rate_per_10k'])
r2_2 = r2v ** 2
print(f'Drug use disorder vs rate: r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f}')

x_range2 = [df['drug_disorder_pct'].min(), df['drug_disorder_pct'].max()]
y_range2 = [slope2 * x + intercept2 for x in x_range2]

fig2 = px.scatter(
    df, x='drug_disorder_pct', y='homeless_rate_per_10k', color='state_postal',
    hover_name='coc_name',
    title=f'Drug Use Disorder Rate (State Proxy) vs. Homeless Rate (CoC Level)<br><sup>r={r2v:.3f}, R²={r2_2:.3f}, p={p2:.4f}</sup>',
    labels={'drug_disorder_pct': 'Drug Disorder % (state proxy)', 'homeless_rate_per_10k': 'Rate per 10k'},
)
fig2.add_trace(go.Scatter(x=x_range2, y=y_range2, mode='lines', name='Regression', line=dict(color='black', dash='dash'), showlegend=False))
fig2.show()
```
