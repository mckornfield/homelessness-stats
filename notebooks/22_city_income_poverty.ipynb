# 22 — Income & Poverty vs. Homeless Rate (City/CoC)

```python
from pathlib import Path
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_city_data.csv')
df_inc = df.dropna(subset=['homeless_rate_per_10k', 'median_income']) if 'median_income' in df.columns else pd.DataFrame()
print(f'CoCs with income data: {len(df_inc)}')
```

```python
if len(df_inc) > 5:
    slope, intercept, r, p, se = stats.linregress(df_inc['median_income'], df_inc['homeless_rate_per_10k'])
    r2 = r ** 2
    x_range = [df_inc['median_income'].min(), df_inc['median_income'].max()]
    y_range = [slope * x + intercept for x in x_range]
    fig = px.scatter(
        df_inc, x='median_income', y='homeless_rate_per_10k', text='city_name',
        color='homeless_rate_per_10k', color_continuous_scale='Reds',
        hover_name='coc_name',
        title=f'Median Income vs. Homeless Rate (City/CoC)<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f}</sup>',
        labels={'median_income': 'Median Household Income ($)', 'homeless_rate_per_10k': 'Rate per 10k'},
    )
    fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
    fig.update_traces(textposition='top center', selector=dict(mode='markers+text'))
    fig.show()
else:
    # FMR as proxy for income
    df2 = df.dropna(subset=['homeless_rate_per_10k', 'median_1br_fmr'])
    slope, intercept, r, p, se = stats.linregress(df2['median_1br_fmr'], df2['homeless_rate_per_10k'])
    r2 = r ** 2
    x_range = [df2['median_1br_fmr'].min(), df2['median_1br_fmr'].max()]
    y_range = [slope * x + intercept for x in x_range]
    fig = px.scatter(
        df2, x='median_1br_fmr', y='homeless_rate_per_10k', text='city_name',
        color='homeless_rate_per_10k', color_continuous_scale='Reds',
        hover_name='coc_name',
        title=f'FMR (income proxy) vs. Homeless Rate (City/CoC)<br><sup>r={r:.3f}, R²={r2:.3f}, p={p:.4f}</sup>',
        labels={'median_1br_fmr': 'Median 1BR FMR ($)', 'homeless_rate_per_10k': 'Rate per 10k'},
    )
    fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', name='Regression', line=dict(color='darkred', dash='dash')))
    fig.update_traces(textposition='top center', selector=dict(mode='markers+text'))
    fig.show()
```
