# 15 — Sheltered vs. Unsheltered Breakdown by CoC

Compares sheltered vs. unsheltered homeless populations across CoCs.
Key question: which CoCs have the highest share of unsheltered homeless?

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

ROOT = Path().resolve().parent
df = pd.read_csv(ROOT / 'data' / 'processed' / 'merged_county_data.csv')
df = df.sort_values('unsheltered_pct', ascending=False)
print(f'Loaded {len(df)} CoCs')
```

```python
top30 = df.nlargest(30, 'total_homeless')[['coc_name', 'sheltered_homeless', 'unsheltered_homeless', 'unsheltered_pct']].copy()
top30 = top30.sort_values('unsheltered_pct', ascending=True)

fig = px.bar(
    top30, x=['sheltered_homeless', 'unsheltered_homeless'], y='coc_name',
    orientation='h', barmode='stack',
    color_discrete_map={'sheltered_homeless': '#4a90d9', 'unsheltered_homeless': '#d94a4a'},
    title='Sheltered vs. Unsheltered Homeless: Top 30 CoCs by Total Count',
    labels={'value': 'Homeless Count', 'coc_name': '', 'variable': 'Status'},
)
fig.update_layout(height=800, legend_title='Status')
fig.for_each_trace(lambda t: t.update(name=t.name.replace('_homeless', '').replace('_', ' ').title()))
fig.show()
```

```python
fig2 = px.bar(
    top30.sort_values('unsheltered_pct', ascending=True),
    x='unsheltered_pct', y='coc_name', orientation='h',
    color='unsheltered_pct', color_continuous_scale='OrRd',
    title='Unsheltered % of Total Homeless: Top 30 CoCs by Total Count',
    labels={'unsheltered_pct': '% Unsheltered', 'coc_name': ''},
)
fig2.update_layout(height=800)
fig2.show()
```

```python
print(f'National unsheltered %: {df["unsheltered_pct"].mean():.1f}%')
print(f'Highest unsheltered %: {df.nlargest(5, "unsheltered_pct")[["coc_name", "unsheltered_pct"]].to_string(index=False)}')
print(f'Lowest unsheltered %: {df.nsmallest(5, "unsheltered_pct")[["coc_name", "unsheltered_pct"]].to_string(index=False)}')
```

## Insights

- **Shelter capacity**: CoCs with better shelter infrastructure have lower unsheltered rates
- **Climate effect**: Mild-weather regions tend to have higher unsheltered rates
- **Service delivery**: The gap shows where shelter beds are under-resourced
- **Policy priority**: Low unsheltered % suggests effective system response
